## Strings

At the heart of many musical instruments is the string - a simple, yet deceptively complex vibrating object, which brings together many of the fundamentals we've already seen. We can conceptually treat the string as a one-dimensional object, i.e., we'll consider the vibration of the string along its length (its "x-direction"). This will greatly simplify the analysis, but will nevertheless provide some deep insight into the mechanism of vibration and the consequence for the sound produced. We'll begin by discussing the acoustics of the string, and then follow this up with a model that uses so-called digital waveguides.
